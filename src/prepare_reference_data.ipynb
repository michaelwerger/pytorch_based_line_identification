```python
def load_images(path):
    train_images = []
    train_labels = []

    dir_index = 0
    for directory in sorted(os.listdir(path)):
        if(directory.startswith('.') == False):

            #print("Buchstabe: {}".format(LIST_OF_CHARS[dir_index]))
            for filename in sorted(os.listdir(os.path.join(path, directory))):

                if(filename.startswith('.') == False):
                        
                        image_path = os.path.join(path, directory,filename    )
                        img =cv2.imread(image_path)
                        img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
                        img = img.reshape([IMAGE_SIZE, IMAGE_SIZE,1])
                        train_images.append(img)
                        train_labels.append(dir_index)
            dir_index = dir_index + 1

    print(len(train_images),train_labels)
    return np.array(train_images)/255, np.array(keras.utils.to_categorical(train_labels,NUM_OUTPUTS))#len(train_labels)))
```

```python
def build_model():
        
    inputs = keras.Input(shape=(IMAGE_SIZE,IMAGE_SIZE,1))
    reshape = keras.layers.Reshape(target_shape=(IMAGE_SIZE,IMAGE_SIZE,1))(inputs)
    conv2d_1 = keras.layers.Conv2D(filters=32, kernel_size=(5,5), activation='relu')(reshape)
    maxpooling2d_1 = keras.layers.MaxPooling2D(pool_size=(2, 2))(conv2d_1)
    conv2d_2 = keras.layers.Conv2D(filters=64, kernel_size=(5,5), activation='relu')(maxpooling2d_1)
    maxpooling2d_2 = keras.layers.MaxPooling2D(pool_size=(2, 2))(conv2d_2)
    flatten = keras.layers.Flatten()(maxpooling2d_2)
    dense_1 = keras.layers.Dense(1024)(flatten)
    dropout = keras.layers.Dropout(0.5)(dense_1)
    outputs = keras.layers.Dense(NUM_OUTPUTS, activation='softmax')(dropout)
    model = keras.Model(inputs=inputs, outputs=outputs)

    return model
```
